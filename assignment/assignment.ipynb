```python
# QUESTION 2 - For the `./data/sharks.csv` data covered in the lecture, clean the "Type" variable as well as you can, and explain the choices you make.
```

```python
df = pd.read_csv('./data/sharks.csv',low_memory=False)
df.head()
```

```python
# df['Type'].value_counts()

type = df['Type']
type.value_counts()
# check for all types, a lot of these are similar... like boat, boating, boatomg?, and sea disaster can all fall under the watercraft category
# also a lot of uncertain/unclean data
```

```python

type = type.replace(['Boat', 'Boating', 'Boatomg', 'Sea Disaster'], 'Watercraft') # combine all water related data to a single watercraft type
type = type.replace(['Invalid', 'Questionable', 'Unconfirmed', 'Unverified', 'Under investigation'], np.nan) # combine all uncertain/unclean data to NaN

type.value_counts()

df['Type'] = type # Replace the 'Type' variable with the cleaned version

df['Type'].value_counts()
# sum(df['Type'].isnull())
```

```python
fatal = df['Fatal (Y/N)']
fatal.value_counts() # very unclean data with the fatal column... there's even random gender/year data in here
```

```python
fatal = fatal.replace(['UNKNOWN', 'F', 'M', '2017'], np.nan) # replace all unclean/unknown data with nan
fatal = fatal.replace('y', 'Y') # replace that lowercase y to capital Y
fatal.value_counts()

df['Fatal (Y/N)'] = fatal
df['Fatal (Y/N)'].value_counts()
```

```python
# QUESTION 4 - Missing values, not at random: For the pretrial data covered in the lecture, clean the ImposedSentenceAllChargeInContactEvent variable 
# as well as you can, and explain the choices you make. (Hint: Look at the SentenceTypeAllChargesAtConvictionInContactEvent variable.)
```

```python
url = 'http://www.vcsc.virginia.gov/pretrialdataproject/October%202017%20Cohort_Virginia%20Pretrial%20Data%20Project_Deidentified%20FINAL%20Update_10272021.csv'
df = pd.read_csv(url,low_memory=False)
df.head()
```

```python
sentence_length = df['ImposedSentenceAllChargeInContactEvent']
sentence_type = df['SentenceTypeAllChargesAtConvictionInContactEvent']
print(sentence_length.unique()) # weird, ' ' is considered a value
```

```python
sentence_length = pd.to_numeric(sentence_length,errors='coerce') # change all data to numeric
print( np.sum(sentence_length.isnull()),'\n') # 9053 null data points...

# According to pdf, cat 4 is dismissed cases and cat 9 is na
sentence_length = sentence_length.mask(sentence_type == 4, 0) # replace cat 4 with 0
sentence_length = sentence_length.mask(sentence_type == 9, np.nan) # replace cat 9 with nan

# Make new dummy variable to lessen the number of NAs
NA_sentence_length = sentence_length.isnull()
# print( pd.crosstab(NA_sentence_length, sentence_type), '\n')
print( np.sum(NA_sentence_length),'\n') # 274 now

df['ImposedSentenceAllChargeInContactEvent'] = sentence_length
```
